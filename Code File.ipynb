## შესავალი და ბიბლიოთეკები

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from matplotlib.colors import ListedColormap

print("ბიბლიოთეკები ჩატვირთულია!")
```

## მონაცემების მომზადება

```python
X = np.array([[0, 0], 
              [0, 1], 
              [1, 0], 
              [1, 1]])

y_dict = {
    "AND": np.array([0, 0, 0, 1]), 
    "OR":  np.array([0, 1, 1, 1]), 
    "XOR": np.array([0, 1, 1, 0])  
}

print("მონაცემები მზადაა.")
```

## მოდელის შექმნა და ვიზუალიზაციის ფუნქცია

```python
def train_and_visualize(logic_gate, y):
    # ვიყენებთ MLPClassifierს (Multi-Layer Perceptron).
    #hidden_layer_sizes=(4,) ერთი ფარული შრე 4 ნეირონით
    model = MLPClassifier(hidden_layer_sizes=(4,), activation='relu', max_iter=3000, random_state=42)
    
    model.fit(X, y)
    score = model.score(X, y)
    print(f"--- {logic_gate} Logic ---")
    print(f"Accuracy: {score * 100:.0f}%")
    print(f"Predictions: {model.predict(X)}")
    
    #ვიზუალიზაცია 
    x_min, x_max = -0.5, 1.5
    y_min, y_max = -0.5, 1.5
    h = 0.02
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.figure(figsize=(6, 4))
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=ListedColormap(['#FFCCCC', '#CCFFCC']))
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', s=200, cmap=ListedColormap(['#FF0000', '#00FF00']))
    plt.title(f"Decision Boundary for {logic_gate} (Hidden Layer Used)")
    plt.xlabel("Input 1")
    plt.ylabel("Input 2")
    plt.grid(True)
    plt.show()

print("ფუნქცია განსაზღვრულია.")
```
